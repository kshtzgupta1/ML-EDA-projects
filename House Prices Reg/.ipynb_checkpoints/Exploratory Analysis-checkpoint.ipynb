# Distributions of categorical features

Next, let's take a look at the distributions of our categorical features.

<br>
Display summary statistics for categorical features.

```python
# Summarize categorical features
df.describe(include = "object")
```

Plot bar plot for the <code style="color:steelblue">'exterior_walls'</code> feature.

```python
# Bar plot for 'exterior_walls'
plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.countplot(y = "exterior_walls", data = df, order = df['exterior_walls'].value_counts().index)
plt.show()
```

```python
# Plot bar plot for each categorical feature
plt.figure(figsize = (10,6))
sns.set(style = 'darkgrid')
for feature in (df.dtypes[df.dtypes == "object"]).index:
    sns.countplot(y = feature, data = df, order = df[feature].value_counts().index)
    plt.show()
    
```

We note that exterior_walls and roof both suffer from sparse classes

<br id="segmentations">
# Segmentations
Let's create some segmentations.
<br>


```python
# Segment tx_price by property_type and plot distributions
plt.figure(figsize = (16,8))
sns.set(style = "whitegrid")
sns.boxplot(y = "property_type", x = 'tx_price', data = df)
```

Segment by <code style="color:steelblue">'property_type'</code> and calculate the average value of each feature within each class:

```python
# Segment by property_type and display the means within each class
df.groupby('property_type').mean()
```

<br><hr style="border-color:royalblue;background-color:royalblue;height:1px;">
## <span style="color:RoyalBlue"></span>

On average, it looks like single family homes are more expensive.

How else do the different property types differ? Let's see:

<br>


```python
plt.figure(figsize = (16,10))
sns.boxplot(y = 'property_type', x = 'sqft', data = df)

# Which type of property is larger, on average? Single Family homes
# Which type of property sees greater variance in sizes? Single family homes
# Does the difference in distributions between classes make intuitive sense? Yes. because 
# apartments/ condos/ townhouse are largely standard living units which don't differ by much
# from each other but single family homes can differ a lot from each other. So, yeah it makes 
# sense
```

<br>


```python
# Segment by property_type and display the means and standard deviations within each class
# So far it seems like group_by is only useful when we are dealing with aggregations but we cannot
# extract out columns, we can only apply aggregate functions on the groupoed object
grouped = df.groupby('property_type')
grouped.agg([np.mean, np.std])
```

# Correlations

Let's take a look at the relationships between **numeric features** and **other numeric features**.

<br>
Create a <code style="color:steelblue">correlations</code> dataframe from <code style="color:steelblue">df</code>.

```python
# Calculate correlations between numeric features
df.corr()
```

Visualize the correlation grid with a heatmap

```python
#Make the figsize 7 x 6. So here we are making a heatmap with seaborn package. and in seaborn 
# we have to adjust the figure size by using plt.figure
plt.figure(figsize = (10,10))
sns.set(style = "darkgrid")
# Plot heatmap of correlationss
sns.heatmap(df.corr(), cbar = False)
 
plt.show()
```

```python
# Change color scheme
sns.set(style = "white")
# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

# Plot heatmap of correlations
sns.heatmap(df.corr())
```

See how the cells for <code style="color:steelblue">'basement'</code> are now white? That's what we want because they were not able to be calculated.

<br>
**Let's display the correlation values in each cell.**


```python
# Make the figsize 10 x 8
correlations = df.corr()
plt.figure(figsize = (14,14))
sns.set(style = "white")
mask = np.zeros_like(correlations, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
# Plot heatmap of annotated correlations
sns.heatmap(correlations * 100, cbar = False, mask = mask, annot = True, fmt = ".0f", cmap= "Greens", square = True)
plt.show()

# 
# sns.heatmap(df, cmap="BuPu")
# sns.heatmap(df, cmap="Greens")

```

<br>


```python
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
# Make the figsize 9 x 8
plt.figure(figsize = (9,8))
correlations = df.corr()
sns.set(style = "white")
mask = np.zeros_like(correlations, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
# Plot heatmap of correlations
sns.heatmap(correlations*100, mask = mask, fmt = ".0f", cbar = False, annot = True, color = "green")
plt.show()
```
