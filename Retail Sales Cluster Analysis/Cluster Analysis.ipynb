### Import libraries and load the datasets.



```python
# NumPy for numerical computing
import numpy as np

# Pandas for DataFrames
import pandas as pd


# Matplotlib for visualization
import matplotlib.pyplot as plt

# display plots in the notebook
%matplotlib inline

# Seaborn for easier visualization
import seaborn as sns

# Scikit-Learn's make_pipeline function
from sklearn.pipeline import make_pipeline

# Scikit-Learn's StandardScaler
from sklearn.preprocessing import StandardScaler

# Scikit-Learn's KMeans algorithm (added later)
from sklearn.cluster import KMeans
```

```python
# Import analytical base table
base_df = pd.read_csv('project_files/analytical-base-table.csv', index_col = 0)

# Import thresholded item features
threshold_item_data = pd.read_csv('project_files/threshold_item_data.csv', index_col = 0)

# Import PCA item features
pca_item_data = pd.read_csv('project_files/pca_item_data.csv', index_col = 0)
```

```python
# Print shape of each dataframe
print(base_df.shape)
print(threshold_item_data.shape)
pca_item_data.shape
```

<span id="k-means"></span>
# 1. K-Means



```python
# Join base_df with threshold_item_data
threshold_df = base_df.join(threshold_item_data)

# Display first 5 rows of threshold_df
threshold_df.head()
```

```python
# Join base_df with pca_item_data
pca_df = base_df.join(pca_item_data)

# Display first 5 rows of pca_df
pca_df.head()
```

<span id="base"></span>
# 2. Base DF

The first set of features we'll consider is **"Base DF"** which only includes the purchase pattern features from our analytical base table.



```python
# First 5 observations of base_df
base_df.head()
```

```python
# K-Means model pipeline
k_means_base_df = make_pipeline(StandardScaler(), KMeans(random_state = 123, n_clusters = 3))
```

```python
# Fit K-Means pipeline
k_means_base_df.fit(base_df)
```

```python
# Save clusters to base_df
base_df['cluster'] = k_means_base_df.predict(base_df)

# Display first 5 rows of base_df
base_df.head()
```

```python
# Scatterplot, colored by cluster
sns.set_style("darkgrid")
sns.lmplot(x='total_sales', y='avg_cart_value', hue = 'cluster', data = base_df, fit_reg = False)
plt.show()
```

<span id="threshold"></span>
# Threshold DF

The second set of features we'll consider is **"Threshold DF"**. 
* This includes the purchase pattern features from our analytical base table.
* This also includes the purchase counts for the 20 most popular items.



```python
# K-Means model pipeline
k_means_threshold_df = make_pipeline(StandardScaler(), KMeans(random_state = 123, n_clusters = 3))
```

```python
# Fit K-Means pipeline
k_means_threshold_df.fit(threshold_df)
```

```python
# Save clusters to threshold_df
threshold_df['cluster'] = k_means_threshold_df.predict(threshold_df)
# Display first 5 rows of threshold_df
threshold_df.head()
```

```python
# Scatterplot, colored by cluster
sns.lmplot(x = 'total_sales', y = 'avg_cart_value', data = threshold_df, fit_reg = False, hue = "cluster")
plt.show()
```

<span id="pca"></span>
# PCA DF

The last set of features we'll consider is **"PCA DF"**. 
* This includes the purchase pattern features from our analytical base table.
* This also includes the first 125 principal component features from the item data.


<br>

```python
# K-Means model pipeline
k_means_pca_df = make_pipeline(StandardScaler(), KMeans(random_state = 123, n_clusters = 3))
```

```python
# Fit K-Means pipeline
k_means_pca_df.fit(pca_df)
```

```python
# Save clusters to pca_df
pca_df['cluster'] = k_means_pca_df.predict(pca_df)

# Display first 5 rows of pca_df
pca_df.head()
```

```python
# Scatterplot, colored by cluster
sns.lmplot(x = 'total_sales', y = 'avg_cart_value',
           data = pca_df, fit_reg = False, hue = 'cluster')
plt.show()
```

<span id="comparison"></span>
# Model comparison

Each of those 3 dataframes have the same indices:

```python
# Check all indices are identical
print( all( base_df.index == threshold_df.index ) )
print( all( base_df.index == pca_df.index) )
```

```python
# Adjusted Rand index
from sklearn.metrics import adjusted_rand_score
```

```python
# Similary between base_df.cluster and threshold_df.cluster

print(adjusted_rand_score(base_df.cluster, threshold_df.cluster))
```

```python
# Similary between threshold_df.cluster and pca_df.cluster
adjusted_rand_score(threshold_df.cluster, pca_df.cluster)
```

```python
# Similary between base_df.cluster and pca_df.cluster
adjusted_rand_score(base_df.cluster, pca_df.cluster)
```

**We notice that compared to pca_df.cluster, threshold_df.cluster is more like the one created from the base_df.**
