#### Correlation Heatmap
Let's plot a correlation heatmap to further investigate the relationship between the numerical features.


```python
# Calculate correlations between numeric features
df.corr()
```

```python
# Plot correlation heatmap

# Make the figsize 9 x 8
plt.figure(figsize = (9,8))
correlations = df.corr()
sns.set(style = "white")
mask = np.zeros_like(correlations, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
# Plot heatmap of correlations
sns.heatmap(correlations*100, mask = mask, fmt = ".0f", cbar = False, annot = True, color = "green")
plt.show()

        
```

There seems to be a weak positive correlation between number of projects and avg_monthly_hours

<span id="categorical"></span>
# Distributions of categorical features

Next, let's take a look at the distributions of our categorical features.



```python
# Summarize categorical features
df.describe(include = "object")
```

```python
# Plot bar plot for each categorical feature
categorical_features = df.select_dtypes(include = 'object').columns.values
plt.figure(figsize = (10, 10))
for feature in categorical_features:
    sns.countplot(y = feature, data = df)
    plt.show()
```

In the countplot for department we notice a class, 'temp' which is irrelevant to our analysis. In addition to sparse classes, there seem to be duplicate classes like 'IT' & 'information management'. We will handle these in data cleaning.

<span id="segmentations"></span>
# Segmentations





```python
# Segment satisfaction by status and plot distributions
plt.figure(figsize = (8,8))
sns.violinplot(y = 'status', x ='satisfaction', data = df)
plt.show()
```

This chart compares of the distributions of satisfaction between employees who've left and those that are still employed.
Those still employed tend to more satisfied, which makes intuitive sense.

```python
# Segment last_evaluation by status and plot distributions
plt.figure(figsize = (8,8))
sns.violinplot(y = 'status', x ='last_evaluation', data = df)
plt.show()
```

This chart compares of the distributions of last_evaluations between employees who've left and those that are still employed. The distribution for those who've left seem a bit peculiar. We will investigate this further later.

```python
# Segment by status and display the means within each class

df.groupby('status').mean()
```

<span id="advanced-segmentations"></span>
# Advanced segmentations

Because our target variable 'status' is categorical, it can often be helpful to conduct segmentation analysis. 



```python
# Scatterplot of satisfaction vs. last_evaluation

plt.figure(figsize = (8,8))
sns.lmplot(x = 'satisfaction',
           y ='last_evaluation', data = df, scatter_kws= {'alpha': 0.2}, hue = 'status', fit_reg = False)

plt.show()
```

This chart is too cluttered.
Instead, it might be more helpful to plot a separate scatterplot just for those who've 'Left'

```python
# Scatterplot of satisfaction vs. last_evaluation, only those who have left
sns.lmplot(x = 'satisfaction',
           y ='last_evaluation', data = df[df.status == 'Left'], fit_reg = False)
plt.show()
```

It looks like there are three clusters of people who've left. Maybe we can explicitly highlight these clusters by creating 3 indicator features during feature engineering:

"Underperformers: for last_evaluation < 0.6

"Unhappy" for satisfaction_level < 0.2... maybe they are overqualified?

"Overachievers" last_evaluation > 0.8 and satisfaction > 0.7
